```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataframe = pd.read_csv("combined_attacks_data.csv")

# Display the first few rows of the dataframe
print(dataframe.head())

# Get a summary of the dataframe
print(dataframe.info())

# Get basic statistics of numeric columns
print(dataframe.describe())

# Get basic statistics of non-numeric columns
print(dataframe.describe(include=[object]))

# Check for missing values
missing_values = dataframe.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Drop columns with a high number of missing values
dataframe = dataframe.drop(columns=['unnamed:_22', 'unnamed:_23'])

# Fill missing values with appropriate values or drop rows with missing values
dataframe['age'].fillna('Unknown', inplace=True)
dataframe.dropna(subset=['latitude', 'longitude'], inplace=True)

# Convert data types if necessary
dataframe['year'] = pd.to_numeric(dataframe['year'], errors='coerce')
dataframe['date'] = pd.to_datetime(dataframe['date'], errors='coerce')

# Plot the distribution of attacks by year
plt.figure(figsize=(12, 6))
sns.histplot(dataframe['year'].dropna(), bins=30, kde=True)
plt.title('Distribution of Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()

# Plot the count of attacks by country
plt.figure(figsize=(12, 6))
sns.countplot(y=dataframe['country'], order=dataframe['country'].value_counts().index, palette='viridis')
plt.title('Count of Attacks by Country')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()

# Plot the distribution of attacks by age
plt.figure(figsize=(12, 6))
sns.histplot(dataframe['age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Attacks by Age')
plt.xlabel('Age')
plt.ylabel('Number of Attacks')
plt.show()

# Plot the count of fatal attacks
plt.figure(figsize=(6, 6))
sns.countplot(x=dataframe['fatal_(y/n)'], palette='coolwarm')
plt.title('Count of Fatal Attacks')
plt.xlabel('Fatal (Y/N)')
plt.ylabel('Number of Attacks')
plt.show()

# Plot a heatmap of missing values
plt.figure(figsize=(12, 6))
sns.heatmap(dataframe.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()
```

```python
import pandas as pd
import folium
from IPython.display import display
                                                                           # Bear attacks
# Load your data
# Replace 'your_data_file.csv' with the path to your actual data file
df = pd.read_csv('combined_attacks_data.csv')

# Filter the data to include only rows with valid latitude and longitude
df = df.dropna(subset=['latitude', 'longitude'])

# Create a base map centered around the mean latitude and longitude
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=2)

# Add points to the map
for idx, row in df.iterrows():
    # Determine the type of species
    species = row['species'] if pd.notna(row['species']) else row['bear_species']
    # Set a different color for each species
    color = 'blue' if species.lower() == 'shark' else 'red'
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f'Species: {species}',
        icon=folium.Icon(color=color)
    ).add_to(m)

# Display the map in Jupyter Notebook
display(m)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data into a Pandas DataFrame
shark_data_raw_df = pd.read_excel("GSAF5-1.xls")

# Drop columns that are not necessary
shark_data_cleaned_df = shark_data_raw_df.drop(columns=['Unnamed: 22', 'Unnamed: 23'])

# Drop rows with missing 'Case Number'
shark_data_cleaned_df = shark_data_cleaned_df.dropna(subset=['Case Number'])

# Drop rows with missing 'Location' or 'Species'
shark_data_cleaned_df = shark_data_cleaned_df.dropna(subset=['Location', 'Species'])

# Aggregate the data by Location and Species
location_species_counts = shark_data_cleaned_df.groupby(['Location', 'Species']).size().reset_index(name='Counts')

# Display the aggregated data
print(location_species_counts.head())

# Create a bar plot using Seaborn
plt.figure(figsize=(14, 8))
sns.barplot(data=location_species_counts, x='Location', y='Counts', hue='Species')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Shark Attack Incidents by Location and Species')
plt.xlabel('Location')
plt.ylabel('Number of Incidents')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
